```python
import numpy as np
from scipy.stats import ttest_ind

# Data
boys = np.array([4, 5, 6, 5])
girls = np.array([6, 7, 8, 7])

# T-test
t_stat, p_value = ttest_ind(boys, girls)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```
